```python
class Orbita(object):
    """
    Esta classe determina a trajetória circular de um objeto com raio e período determinado.
    Sendo t = tempo ; R = raio planeta ; P = período Planeta ; r = raio satélite ; p = período satélite. 
    """
    def __init__(self,t,R,P,r,p,xp=0,yp=0,xs=0,ys=0):
        self.t = t
        self.R = R
        self.P = P
        self.r = r
        self.p = p
        self.xp = xp
        self.yp = xp
        self.xs = xs
        self.ys = ys
        print('Objeto Criado')
        
    def pos_planeta(self):
        """
        Esta função determina a posição X e Y da órbita do planeta em função do tempo em anos e de raio medio orbital em unidades astronômicas
        """
        self.xp = self.R*np.cos(2*self.t*np.pi/self.P)
        self.yp = self.R*np.sin(2*self.t*np.pi/self.P)
        return print('Posições X e Y em função do tempo (anos) respectivamente para o planeta'),self.xp, self.yp
    
    def pos_satelite(self):
        """
        Esta função determina a posição X e Y da órbita de um satélite em função do tempo em anos e de raio medio orbital em unidades astronômicas
        """
        self.xs = self.xp + (self.r*np.cos(2*self.t*np.pi/self.p))
        self.ys = self.yp + (self.r*np.sin(2*self.t*np.pi/self.p))
        return print('Posições X e Y em função do tempo (anos) respectivamente para o satélite'),self.xs, self.ys
    

    def Grafico_planeta(self):
        """
        Esta função grafica a órbita do planeta em termos de suas posições no tempo
        """
        Gráfico_xy = plt.subplot(2,2,1), plt.plot(self.xp,self.yp,label=(r'X $vs$ Y em função do tempo')),plt.legend(loc=3), plt.title("Órbita do Planeta"),plt.legend(loc=9),plt.xlabel('Coordenada X(ua)'), plt.ylabel('Coordenada Y(ua)');
        Gráfico_tempo = plt.subplot(2,2,2), plt.plot(self.t,self.xp,label=r'X de t'),plt.plot(self.t,self.yp,label=r'Y de t'),plt.legend(loc=3), plt.xlabel("Tempo (anos)"), plt.ylabel("Posição()"),plt.title('Posições em função do tempo');
        return Gráfico_xy, Gráfico_tempo
    
    def Grafico_satelite(self):
        """
        Esta função grafica a órbita do satélite em termos de suas posições no tempo
        """
        Gráfico_xy = plt.subplot(2,2,1), plt.plot(self.xs,self.ys,label=(r'X $vs$ Y em função do tempo')),plt.legend(loc=3), plt.title("Órbita do satélite"),plt.xlabel('Coordenada X(ua)'), plt.ylabel('Coordenada Y(ua)');
        Gráfico_tempo = plt.subplot(2,2,2), plt.plot(self.t,self.xs,label=r'X de t'),plt.plot(self.t,self.ys,label=r'Y de t'),plt.legend(loc=3), plt.xlabel("Tempo (anos)"), plt.ylabel("Posição()"),plt.title('Posições((x,y)[ua]) em função do tempo(anos)');
        return Gráfico_xy, Gráfico_tempo
    
    def Grafico_orbita(self):
        """
        Esta função grafica a órbita do planeta e do satélite associado em termos de suas posições no tempo
        """
        return plt.plot(self.xp,self.yp,label='Trajetória do planeta'), plt.plot(self.xs,self.ys,label='Trajetória do satélite'),plt.legend(loc=3),plt.title('Órbitas combinadas do planeta e do satélite'),plt.xlabel('Coordenada X(ua)'), plt.ylabel('Coordenada Y(ua)')
```

#### 2. (a) 10% -- Escreva as equações de movimento de um corpo (segunda lei de Newton) em duas dimensões. 

$$F_x = \dfrac{dp_x}{dt}$$

$$\int_{t0}^{t1} F_x dt = \Delta p_x$$

$$\Delta p_x = F_x \Delta t$$

$$F_x = \frac{m v_{x1} - m v_{x0}}{t_{1} - t_{0}} = m \frac{\Delta v_x}{\Delta t}$$

$$F_x = m a_x$$

Para no caso onde existam mais forças atuando no sentido do eixo $x$ temos

$$F_{res,x} = \sum_{n=1}^{N} m a_{xn}$$

onde N é o número de forças atuantes.


Para o eixo $y$ temos:

$$F_y = \frac{dp_y}{dt}$$

$$\int_{t0}^{t1} F_y dt = \Delta p_y$$

$$F_y \Delta t = \Delta p_y$$

$$F_y = \frac{m v_{y1} - m v_{y0}}{t_{1} - t_{0}} = m \frac{\Delta v_y}{\Delta t}$$

$$F_y = m a_y$$

Se existirem mais forças atuando utilizamos o somatório anterior para determinar a força resultante aplicado ao eixo y. Para esse caso, sob interação com um campo gravitacional podemos especificar uma das forças aplicadas sendo a força $peso$. A resultante será portanto:

$$F_{res,y} = m g + \sum_{n=1}^{N} m a_{yn}$$

que poderia ser inclusa ao somatório resultando em:

$$F_{res,y} = \sum_{n=1}^{N} m a_{yn}$$

#### Caso o corpo tenha posição e velocidade iniciais dadas por ($x_0,y_0$), ($v_{x_0},v_{y_0}$), e acelerações dadas por ($a_{x},a_{y}$), escreva a solução geral das equações de movimento, isto é $x$ e $y$ como função de $t$.

Seja uma aceleração $a_x$ e $a_x = \frac{dv_x}{dt}$, com o tempo variando de $0$ à $t$, obtemos a equação para a velocidade de um corpo como:

$$\int_{v_{0x}}^{v_{x}}dv_x = a_x \int_{0}^{t} dt$$

$$v_x = v_{0_x} + a_x t$$

Temos portanto a equação horária da velocidade.

Com isso podemos determinar a equação da posição para este corpo, com $t$ variando de $0$ a $t$, onde $v_x = \frac{dx}{dt}$ variando de $x_0$ a $x$:

$$\int_{x_0}^{x}dx = \int_{0}^{t}(v_{0_x} + a_x t) dt$$

$$x - x_0 = v_{0x} t + \frac{a_x t^2}{2}$$

Determinamos assim a equação que descreve o Movimento Retilíneo Uniformemente Variado:

$$x = x_0 + v_{0_x}t + \frac{a_x t^2}{2}$$

As mesmas equações podem descrever o movimento no eixo x. Assim:

$$\int_{v_{0_y}}^{v_{y}}dv_y = a_y \int_{0}^{t} dt$$

$$v_y = v_{0_y} + a_y t$$

$$\int_{y_0}^{y}dy = \int_{0}^{t}(v_{0_y} + a_y t) dt$$

$$y - y_0 = v_{0_y} t + \frac{a_y t^2}{2}$$

$$y = y_0 + v_{0_y}t + \frac{a_y t^2}{2}$$

####   (b) 20%  -- Escreva uma clase que permita inicializar uma partícula com sua posição e velocidade iniciais, que tenha como instâncias as posições $x$ e $y$ em um tempo $t$. Experimente com os seguintes casos:
*  $(x_0,y_0) = (0,0)$;  $(v_{x0},v_{y0}) = (3,4)$; $a_x = 0$, $a_y = -9.8$.
*  $(x_0,y_0) = (0,0)$;  $(v_{x0},v_{y0}) = (5,5)$; $a_x = 0$, $a_y = -9.8$.

```python
class MRUV(object):
    
    def __init__(self,posicao=[0,0],velocidade=[0,0],aceleracao=[0,0],tempo=0,theta=0,Vetor_velocidade=0,x=0,y=0,vx=0,vy=0,hmax=0,xmax=0,tmax=0,g=9.8):
        self.x0 = posicao[0]
        self.y0 = posicao[1]
        self.v0x = velocidade[0]
        self.v0y = velocidade[1]
        self.ax = aceleracao[0]
        self.ay = aceleracao[1]
        self.t = tempo
        self.theta = theta
        self.Vel = Vetor_velocidade
        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy
        self.hmax = hmax
        self.xmax = xmax
        self.tmax = tmax
        self.g = g
        print('OBJETO CRIADO')
        
    def pos(self):
        """
        Função que calcula a posição de um corpo com posição inicial (x,y), velocidade inicial (Vox,Voy), aceleração constante (Aox,Aoy) em um intervalo de tempo, para um sistema de unidades único
        """
        self.x = (self.x0) + (self.v0x * (self.t)) + (0.5 * self.ax * ((self.t)**2))
        self.y = (self.y0) +( self.v0y * (self.t)) + (0.5 * self.ay * ((self.t)**2))
        return self.x , self.y
    
    def vel(self):
        """
        Função que calcula a velocidade de um corpo com velocidade inicial (Vox,Vox) e aceleração constante (Aox,Aoy) em função do tempo.
        """
        self.vx = self.v0x + (self.ax * (self.t))
        self.vy = self.v0y + (self.ay * (self.t))
        return self.vx , self.vy
    
    def ace(self):
        """
        Função que fornece a aceleração dada, contante.
        """
        self.ax = self.ax + 0*self.t
        self.ay = self.ay + 0*self.t
        return self.ax, self.ay
    
    def h_max(self):
        """
        Função que calcula a altura máxima que um corpo pode atindir sob uma aceleracao negativa em y (gravidade, por exemplo) em função do seu ângulo de lançamento
        """
        self.hmax = ((self.Vel * np.sin(self.theta))**2)/(2*self.g)
        return self.hmax
        
    def x_max(self):
        """
        Função que calcula o alcance máximo em x em função do seu ângulo de lançamento
        """
        self.xmax = (((self.Vel)**2)*np.sin(2*self.theta))/self.g
        return self.xmax
        
    def tempo_subida(self):
        """
        Função que calcula o tempo de subida de um corpo em função de seu ângulo de lançamento
        """
        self.tmax = 2*(self.Vel * np.sin(self.theta))/self.g
        return self.tmax
    
    def g_hmax(self):
        """
        Função que grafica a altura máxima atingida por um corpo ao ser lançado de uma determinada inclinação 
        """
        g_th = plt.plot(self.theta,self.hmax,'k',lw=2), plt.title(r'Gráfico da altura máxima de um projéti em função do ângulo de lançamento $\theta$'), plt.xlabel(r'$\theta (rad)$'), plt.ylabel(r'$y (m)$')
        return g_th
    
    def g_xmax(self):
        """
        Função que grafica o alcance máxim0 atingido por um corpo ao ser lançado de uma determinada inclinação 
        """
        g_th = plt.plot(self.theta,self.xmax,'k',lw=2), plt.title(r'Gráfico do ancance máximo de um projétil em função do ângulo de lançamento $\theta$'), plt.xlabel(r'$\theta (rad)$'), plt.ylabel(r'$x (m)$')
        return g_th
    
    def g_tmax(self):
        """
        Função que grafica o tempo de subida de um corpo ao ser lançado de uma determinada inclinação 
        """
        g_th = plt.plot(self.theta,self.tmax,'k',lw=2), plt.title(r'Gráfico do Tempo de subida de um projétil em função do ângulo de lançamento $\theta$'), plt.xlabel(r'$\theta (rad)$'), plt.ylabel(r'$t(s)$')
        return g_th

    def trajetória(self):
        tempos = []
        theta = np.linspace(5*np.pi/180,85*np.pi/180,5)
        for i in range(0,10):
            tempos.append(self.tmax[5*i])
        for j in range(len(tempos)):
            tempo = np.linspace(0,tempos[j],5)
        tempos
        for t in tempos:
            for a in theta:
                temp = np.linspace(0,t)
                x = (self.Vel * 2 * temp * np.cos(a))
                y = (self.Vel * 2 * temp * np.sin(a)) - (0.5 * self.g * ((2 * temp)**2))
                g = plt.plot(x,y,lw=0.5,label='Churros'),plt.xlim(0,12),plt.ylim(0,5.5), plt.title('Gráfico das coordenadas da trajetória de um corpo em MRUV'),plt.xlabel('Alcance x (m)'), plt.ylabel('Altura y (m)')
        
    def g_pos(self):
        """
        Função que grafica a posição de um corpo em termos de suas coordenadas (x,y) e de suas posiçoes individuais em função do tempo.
        """
        graf_p = plt.subplot(2,2,1), plt.plot(self.x,self.y,'k',lw=2), plt.title('Coordenadas do corpo em duas dimensões dada em função do tempo'), plt.xlabel('Coordenada X (m)'), plt.ylabel('Coordenada Y (m)')
        graf_t = plt.subplot(2,2,2), plt.plot(self.t,self.x,'b',lw=2,label='x de t'), plt.plot(self.t,self.y,'r',lw=2,label='y de t'),plt.legend(loc=3), plt.title('Posições x e y para um tempo t'), plt.xlabel('Tempo (s)'), plt.ylabel('Posições em x e y (m)')
        return graf_p , graf_t
    
    def g_vel(self):
        """
        Função que grafica as velocidades (Vx,Vy) de um corpo em função do tempo
        """
        graf_t = plt.plot(self.t,self.vx,'b',lw=2,label=r'$v_x$ de t'),plt.plot(self.t,self.vy,'r',lw=2,label=r'$v_y$ de t'), plt.legend(loc=3), plt.title('Gráfico da velocidade de um corpo em função do tempo'), plt.xlabel('Tempo (s)'), plt.ylabel(r'Velocidades em x e y $(\frac{m}{s}$)')
        return graf_t
    
    def g_ace(self):
        """Função que grafica a aceleração (Ax, Ay) de um corpo em função do tempo"""
        graf_t = plt.plot(self.t,self.ax,'b',lw=2,label=(r'$a_x$ de t')),plt.title('Gráfico da aceleração em função do tempo'),plt.plot(self.t,self.ay,'r',lw=2,label=(r'$a_y$ de t')), plt.legend(loc=6), plt.xlabel('Tempo (s)'), plt.ylabel(r'Acelerações em x e y $(\frac{m}{s^2})$')
        return graf_t
        
        
    
```
