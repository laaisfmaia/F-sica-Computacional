```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# valor exato de f(x)
def f(x):
    return np.exp(x)

# valor exato de f'(x)
def df1(x):
    return np.exp(x)

# valor exato de f"(x)
def df2(x):
    return np.exp(x)


# funo que retorna a primeira derivada, DFC
def Df(f,x):
    dx = x[1]-x[0]
    xp = x+dx
    xm = x-dx
    return (f(xp)-f(xm))/(2*dx)

# funo que retorna a segunda derivada, DFC
def D2f(f,x):
    dx = x[1]-x[0]
    xp = x+dx
    xm = x-dx
    return (f(xp)-2.*f(x) + f(xm))/(dx*dx)

# funo erro que retorna o logaritmo desse erro
def erro_relativo(calculado,exato):
    erro_relativo = (np.abs(calculado - exato))/(np.abs(exato))
    log_erro = np.log(erro_relativo)
    return log_erro
    

def X(): 
    listaDeMaioresValores = []
    for i in range(2, 21):
        x = np.linspace(0,10,2**i)
        tamanhoX = len(x)
        ultimoValorX = x[tamanhoX - 1]
        penultimoValorX = x[tamanhoX - 2]
        dx = ultimoValorX - penultimoValorX
        listaDeMaioresValores.append(dx)        
    return listaDeMaioresValores

x = np.linspace(0,2,19)
func = f(x)
log_erro = erro_relativo(D2f(f,x),np.exp(x))


plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(x,func, label=("f(x)")) , plt.legend(loc=2)
plt.plot(x,Df(f,x),'o', label=("1 derivada de f(x)")) , plt.legend(loc=2)
plt.plot(x,D2f(f,x),'--', label=("2 derivada de f(x)")) , plt.legend(loc=2)
plt.xlabel("x")
plt.ylabel("f(x), Df(x) e D2f(x)")

#Acredito que o grfico esteja incorreto, mas no consegui encontrar o erro. 

plt.subplot(2,1,2)
a = X()
plt.plot(a, log_erro)
plt.title(" $\Delta x$ vs $\log(\epsilon$)")
plt.ylabel("$\Delta x$")
plt.xlabel("$\log(\epsilon$)")
```
