```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# valor exato de f(x)
def f(x):
    return np.exp(x)

# valor exato de f'(x)
def df1(x):
    return np.exp(x)

# valor exato de f"(x)
def df2(x):
    return np.exp(x)


# funo que retorna a primeira derivada, DFC
def Df(f,x):
    dx = x[1]-x[0]
    xp = x+dx
    xm = x-dx
    return (f(xp)-f(xm))/(2*dx)

# funo que retorna a segunda derivada, DFC
def D2f(f,x):
    dx = x[1]-x[0]
    xp = x+dx
    xm = x-dx
    return (f(xp)-2.*f(x) + f(xm))/(dx*dx)

# funo erro que retorna o logaritmo desse erro
def erro_relativo(calculado,exato):
    erro_relativo = (np.abs(calculado - exato))/(np.abs(exato))
    log_erro = np.log(erro_relativo)
    return log_erro
    
def X(dx): 
    listaDeMaioresValores = []
    for i in range(2, 21):
        x = np.linspace(0,10,2**i)
    dx = x[1]-x[0]
    tamanhoX = len(x)
    ultimoValorX = x[tamanhoX - 1]
    listaDeMaioresValores.append(ultimoValorXm)
    
#O objetivo do grfico  mostrar o menor valor de dx para que o erro seja mnimo (?)


func = f(x)
log_erro = erro_relativo(D2f(f,x),np.exp(x))


plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(x,func, label=("f(x)")) , plt.legend(loc=2)
plt.plot(x,Df(f,x),'o', label=("1 derivada de f(x)")) , plt.legend(loc=2)
plt.plot(x,D2f(f,x),'--', label=("2 derivada de f(x)")) , plt.legend(loc=2)
plt.xlabel("x")
plt.ylabel("f(x), Df(x) e D2f(x)")

plt.subplot(2,1,2)
plt.plot(x,log_erro)
plt.title(" $\log(\Delta x)$ vs $\log(\epsilon$)")
plt.ylabel("$\log(\Delta x)$")
plt.xlabel("$\log(\epsilon$)")
```
