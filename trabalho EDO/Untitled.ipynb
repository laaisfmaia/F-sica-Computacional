```python
def pend(y, t, Q, A, w):
    theta, v = y
    dydt = [v, -v/Q - np.sin(theta) + A*np.cos(w*t)]
    return dydt

Q = 10.
w = 2./3.
A = 0.

angle=10
angle = angle*np.pi/180.

y0 = [angle, 0.0]

t = np.linspace(0, 10.*np.pi, 1000)

sol = si.odeint(pend, y0, t, args=(Q,A,w))

plt.plot(t, sol[:, 0], 'b', label='theta(t)')
#plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
